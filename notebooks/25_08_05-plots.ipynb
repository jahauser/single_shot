```julia
using JLD2
using Plots
```

```julia
data = jldopen("25_08_05.jld2", "r")
MVdata = data["MVdata"]
TOOMdata = data["TOOMdata"]
MWPMdata = data["MWPMdata"];
```

# p = q

```julia
plt = plot()

for L in 4:4:12
    T = L^3
    xs = 0.0:0.01:0.1
    ys = [MVdata[(L, T, x, x)][1][10L] for x in xs]
    plot!(plt, xs, ys, label="L=$L", m=3)
end
plot!(plt, xlabel="p=q", ylabel="M", title="MV")
```

```julia
plt = plot()

for L in 4:4:12
    T = L^3
    xs = 0.0:0.01:0.1
    ys = [TOOMdata[(L, T, x, x)][1][end] for x in xs]
    plot!(plt, xs, ys, label="L=$L", m=3)
end
plot!(plt, xlabel="p=q", ylabel="M", title="Toom")
```

```julia
plt = plot()

for L in 4:4:12
    T = L^3
    xs = 0.0:0.01:0.1
    ys = [MWPMdata[(L, T, x, x)][1][L] for x in xs]
    plot!(plt, xs, ys, label="L=$L", m=3)
end
plot!(plt, xlabel="p=q", ylabel="M", title="MWPM")
```
