```julia
q = 0.1
γ = 2.0
ν = 1.0
pc = 0.18
d  = 2

peff(p, q) = q+2p*(1-p)*(1-2q)
time(L) = 10L

plt = plot(
    xlabel = L"(p-p_c)L^{1/\nu}",
    ylabel = L"\chi L^{-\gamma/\nu}"
)
for L in 8:8:24
    # extract the actually-available p values
    ps = sort(unique([
        key[4] for key in keys(data)
        if key[2] == L &&
           key[5] == q &&
           key[6] == peff(key[4], q)
    ]))
    M1s = [
        data[("adv_MWPM", L, 10L, p, q, peff(p,q))][1][time(L)]
        for p in ps
    ]
    M2s = [
        data[("adv_MWPM", L, 10L, p, q, peff(p,q))][2][time(L)]
        for p in ps
    ]
    χ = L^d .* (M2s .- M1s.^2)

    x = (ps .- pc) .* L^(1/ν)
    y = χ .* L^(-γ/ν)

    plot!(plt, x, y;
        label = "L=$L",
        m = 3,
        seriestype = :scatter,
        xlim = (-2, 2)
    )
end

plt
```
