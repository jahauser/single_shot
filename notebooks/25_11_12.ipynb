```julia
algo = "adv_MWPM"
p = 0.05
q = 0.15

plt = plot()
for L in 4:4:16
    xs = q:0.05:0.5
    ys = [data[(algo, L, 10L, p, q, x)][1][L] for x in xs]
    plot!(xs, ys, label="L = $L", xlabel="peff", ylabel="<m> at time L", ylim=(0,0.55), title="steady state <m> with p = $p, q = $q", m=3)
end

plt
```

```julia
algo = "adv_MWPM"
p = 0.05
q = 0.15

plt = plot()
for L in 4:4:16
    xs = q:0.05:0.5
    ys = [data[(algo, L, 10L, p, q, x)][2][L] for x in xs] - [data[(algo, L, 10L, p, q, x)][1][L] for x in xs] .^ 2
    plot!(xs, L^2*ys, label="L = $L", xlabel="peff", ylabel="(<M²> - <M>²)/V at time L", m=3, title="steady state χ with p = $p, q = $q")
end

plt
```
