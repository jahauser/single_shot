```julia
L = 16
p = 0.09
n = L^2

# xs = [mean(noiselayer(initialstate(L), p)) for _ in 1:1000000];

log_binom(n, k) = lgamma(n+1) - lgamma(k+1) - lgamma(n-k+1)

xs = data[(L, 20L, p, p)][7][:, 20]                # assumed x in [0,1]

h = fit(Histogram, xs, nbins=100) 
edges  = h.edges[1]
counts = Float64.(h.weights)
centers = (edges[1:end-1] .+ edges[2:end]) ./ 2
scaled = counts
scaled = -[log(counts[i]) - log_binom(n, centers[i] * n)
            for i in eachindex(centers)]
          
bar(centers, scaled,
    bar_width = step(edges),
    xlabel    = "x",
    ylabel    = "count / C(n, x·n)",
    legend    = false)
```
